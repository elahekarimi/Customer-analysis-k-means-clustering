```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stopwords.words()
```

```python
def message_cleaning(message):
  Test_punc_removed = [i for i in message if i not in string.punctuation]
  Test_punc_removed_join = ''.join(Test_punc_removed)
  Test_punc_removed_clean = [word for word in Test_punc_removed_join.split() if word.lower() not in stopwords.words('english')]
  return Test_punc_removed_clean
```

```python
reviws_clean = reviews_df['verified_reviews'].apply(message_cleaning)
```

```python
print(reviws_clean[3])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(analyzer= message_cleaning)
reviews_countvectorizer = vectorizer.fit_transform(reviews_df['verified_reviews'])
```

```python
print(vectorizer.get_feature_names_out())
```

```python
X = reviews_df.drop(['feedback'], axis=1)
```

```python
y = reviews_df['feedback']
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
x_train.columns = x_train.columns.astype(str)
NB_classifier = MultinomialNB()
NB_classifier.fit(x_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
y_predict_train = NB_classifier.predict(x_train)
y_predict_train
```

```python
cm = confusion_matrix(y_train, y_predict_train)
sns.heatmap(cm, annot = True)
```

```python
x_test.columns = x_test.columns.astype(str)
y_predict_test = NB_classifier.predict(x_test)
cm = confusion_matrix(y_train, y_predict_train)
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(y_test, y_predict_test))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
y_pre = model.predict(x_test)
```

```python
print("Accuracy = {} %".format(100 * accuracy_score(y_pre, y_test)))
```

```python
cm = confusion_matrix(y_pre, y_test)
sns.heatmap(cm, annot= True)
```
